```python
import pandas as pd
from tabulate import tabulate

class Nodo:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class ListaDoblementeEnlazada:
    def __init__(self):
        self.head = None
        self.tail = None

    def insertar(self, data):
        nuevo_nodo = Nodo(data)
        if self.head is None:
            self.head = nuevo_nodo
            self.tail = nuevo_nodo
        else:
            self.tail.next = nuevo_nodo
            nuevo_nodo.prev = self.tail
            self.tail = nuevo_nodo

    def imprimir_lista(self, columnas):
        print('\t'.join(columnas))

        nodo_actual = self.head
        while nodo_actual:
            valores = [str(nodo_actual.data.get(columna, '')) for columna in columnas]
            print('\t'.join(valores))
            nodo_actual = nodo_actual.next

df = pd.read_excel('estaciones-de-transmilenio.xlsx')

# Obtener las columnas del DataFrame
columnas = df.columns.tolist()

# Crear la lista doblemente enlazada
lista = ListaDoblementeEnlazada()
for index, row in df.iterrows():
    lista.insertar(row.to_dict())

# Crear una lista de listas para tabulate
datos_tabla = []
nodo_actual = lista.head
while nodo_actual:
    valores = [nodo_actual.data.get(columna, '') for columna in columnas]
    datos_tabla.append(valores)
    nodo_actual = nodo_actual.next

# Imprimir la tabla con tabulate
print(tabulate(datos_tabla, headers=columnas, tablefmt="grid"))
```

```python
def insertar_dato_extraido(lista, datos_extraidos):

  lista.insertar(datos_extraidos)

df = pd.read_excel('estaciones-de-transmilenio.xlsx')

# Obtener las columnas del DataFrame
columnas = df.columns.tolist()

# Crear la lista doblemente enlazada
lista = ListaDoblementeEnlazada()
for index, row in df.iterrows():
    lista.insertar(row.to_dict())

# Insertar el nuevo dato y guardar la referencia
datos_extraidos = {'Nombre': 'Estación Extraída', 'OBJECTID': 12345, 'Longitud': -74.123, 'Latitud': 4.567}
insertar_dato_extraido(lista, datos_extraidos)
nodo_insertado = lista.tail

# Imprimir el dato insertado con la función insertar_dato_extraido
print("\nDato insertado con insertar_dato_extraido:")
valores_insertado = [nodo_insertado.data.get(columna, '') for columna in columnas]
print(tabulate([valores_insertado], headers=columnas, tablefmt="grid"))
```

```python
def mostrar_datos_adelante_atras(lista, columnas):
    datos_tabla = []
    nodo_actual = lista.tail  # Empezar desde la cola (último nodo insertado)

    while nodo_actual:
        valores = [nodo_actual.data.get(columna, '') for columna in columnas]
        datos_tabla.append(valores)
        nodo_actual = nodo_actual.prev  # Moverse al nodo anterior

    print(tabulate(datos_tabla, headers=columnas, tablefmt="grid"))

# Mostrar los datos de adelante hacia atrás
print("\nDatos de adelante hacia atrás:")
mostrar_datos_adelante_atras(lista, columnas)
```

```python
def mostrar_datos_atras_adelante(lista, columnas):
    datos_tabla = []
    nodo_actual = lista.head  # Empezar desde la cabeza (primer nodo)

    while nodo_actual:
        valores = [nodo_actual.data.get(columna, '') for columna in columnas]
        datos_tabla.append(valores)
        nodo_actual = nodo_actual.next  # Moverse al siguiente nodo

    print(tabulate(datos_tabla, headers=columnas, tablefmt="grid"))

# Mostrar los datos de atrás hacia adelante
print("\nDatos de atrás hacia adelante:")
mostrar_datos_atras_adelante(lista, columnas)
```

```python
# Leer el archivo Excel
df = pd.read_excel('estaciones-de-transmilenio.xlsx')

# Obtener las columnas del DataFrame
columnas = df.columns.tolist()

# Crear la lista doblemente enlazada
lista = ListaDoblementeEnlazada()
for index, row in df.iterrows():
    lista.insertar(row.to_dict())

# Ordenar los primeros 10 datos por OBJECTID
primeros_10_nodos = []
nodo_actual = lista.head
contador = 0
while nodo_actual and contador < 10:
    primeros_10_nodos.append(nodo_actual)
    nodo_actual = nodo_actual.next
    contador += 1

primeros_10_nodos_ordenados = sorted(primeros_10_nodos, key=lambda nodo: nodo.data['OBJECTID'])

# Crear una nueva lista con los datos ordenados
lista_ordenada = ListaDoblementeEnlazada()
for nodo in primeros_10_nodos_ordenados:
    lista_ordenada.insertar(nodo.data)

# Mostrar los datos ordenados
print("\nPrimeros 10 datos ordenados por OBJECTID:")
mostrar_datos_atras_adelante(lista_ordenada, columnas)
```

```python
# Ordenar los datos por nombre de forma descendente
todos_los_nodos = []
nodo_actual = lista.head
while nodo_actual:
    todos_los_nodos.append(nodo_actual)
    nodo_actual = nodo_actual.next

todos_los_nodos_ordenados = sorted(todos_los_nodos, key=lambda nodo: nodo.data['Name'], reverse=True)

# Crear una nueva lista con los datos ordenados
lista_ordenada_nombre = ListaDoblementeEnlazada()
for nodo in todos_los_nodos_ordenados:
    lista_ordenada_nombre.insertar(nodo.data)

# Mostrar los datos ordenados por nombre
print("\nDatos ordenados por nombre (descendente):")
mostrar_datos_atras_adelante(lista_ordenada_nombre, columnas)
```
