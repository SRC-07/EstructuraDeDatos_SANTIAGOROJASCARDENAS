Datos abiertos de: Aplicacion de dosis de vacuna contra COVID-19

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '/content/Aplicaci_n_de_dosis_de_vacuna_contra_COVID-19_20241125.csv'
data = pd.read_csv(file_path)

# Agrupar por territorio y sumar la cantidad de dosis aplicadas
grouped_data = data.groupby('Nom_Territorio')['Cantidad_dosis_aplicadas'].sum().sort_values(ascending=False)

# Crear la gr√°fica
plt.figure(figsize=(12, 8))
grouped_data.plot(kind='bar', color='skyblue', edgecolor='black')

# Agregar valores sobre las barras
for index, value in enumerate(grouped_data):
    plt.text(index, value + 0.01 * value, f'{int(value):,}', ha='center', va='bottom', fontsize=8)

# Configurar etiquetas y t√≠tulo
plt.title('Total de dosis aplicadas por territorio', fontsize=16)
plt.xlabel('Territorio', fontsize=12)
plt.ylabel('Cantidad de dosis aplicadas', fontsize=12)
plt.xticks(rotation=90, fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Mostrar la gr√°fica
plt.show()
```
